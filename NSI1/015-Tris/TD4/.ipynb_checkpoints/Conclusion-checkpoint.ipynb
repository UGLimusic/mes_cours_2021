```python
nb_opel = 0
```

## Tri par sélection

```python
def tri_selection(lst):
    global nb_opel
    n = len(lst)
    for i in range(0,n-1):
        p = indice_minimum(lst,i)
        nb_opel += 2
        lst[i],lst[p]=lst[p],lst[i]
        
def indice_minimum(lst : list, i : int) -> int:
    global nb_opel
    n = len(lst)
    mini = lst[i]
    nb_opel+=1
    i_mini = i
    for j in range(i,n):
        nb_opel +=1
        if lst[j] < mini:
            mini = lst[j]
            nb_opel+=1
            i_mini = j
    return i_mini
```

# Tri par insertion

```python
def tri_insertion(lst: list) -> None:
    n = len(lst)
    for i in range(1,n):
        place(lst,i)
        
def place(lst : list, i : int) -> None:
    global nb_opel
    while i>0 and lst[i]< lst[i-1]:
        nb_opel +=4
        lst[i], lst[i-1] = lst[i-1], lst[i]
        i -= 1
```

# Tri à bulles optimisé

```python
def tri_bulles2(lst : list) -> None:
    global nb_opel
    n = len(lst)
    echange = True
    p = 0
    while p<=n-1 and echange:
        echange = False
        for i in range(n-1-p):
            nb_opel += 2
            if lst[i] > lst[i+1]:
                nb_opel += 2
                lst[i], lst[i+1] = lst[i+1], lst[i]
                echange = True
        p += 1
```

# Comparaisons des complexités

## Sur listes aléatoires

```python
import matplotlib.pyplot as plt
from random import randint
X = []
Y = []
Z = []
T = []
for i in range(1,400):
    L = [randint(1,10000) for _ in range(i)]
    L2 = L[:]
    L3 = L[:]
    X.append(i)
    nb_opel = 0
    tri_selection(L)
    Y.append(nb_opel)
    nb_opel = 0
    tri_insertion(L2)
    Z.append(nb_opel)
    nb_opel = 0
    tri_bulles2(L3)
    T.append(nb_opel)
plt.plot(X,Y,label = "tri par sélection")
plt.plot(X,Z,label = "tri par insertion")
plt.plot(X,T,label = "tri à bulles optimisé")
plt.title("Complexité des tris")
plt.legend()
plt.show()
```
