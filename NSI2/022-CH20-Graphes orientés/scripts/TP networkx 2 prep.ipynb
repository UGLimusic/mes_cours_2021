```python
import matplotlib.pyplot as plt
import networkx as nx
from random import randint, sample


G = nx.Graph()
G.add_edges_from([('C', 'O'), ('C', 'J'), ('C', 'G'), ('C', 'R'), ('O', 'J'), ('O', 'X'), ('J', 'G'), ('J', 'R'), ('J', 'X'), ('G', 'Q'), ('R', 'X'), ('R', 'Q')])
positions = {'C': [-0.31934877, -0.15269138], 'O': [-0.62606592,  0.81546769], 'J': [-0.26602134,  0.25992198], 'G':
    [-0.1577814 , -0.82012323], 'R': [ 0.49072014, -0.0064218 ], 'X': [0.18605347, 0.90384673], 'Q': [ 0.69244381, -1.        ]}

#color_dict={'C': 0, 'O': 1, 'J': 2, 'G': 1, 'R': 1, 'X': 0, 'Q': 0}
color_dict={'C': '#FF0000', 'O': '#00FF00', 'J': '#0000FF', 'G': '#00FF00', 'R': '#00FF00', 'X': '#FF0000', 'Q': '#FF0000'}


color_list = []
for u in G.nodes:
    color_list.append(color_dict[u])

plt.clf()

nx.draw(G, pos=positions,with_labels=True,node_color=color_list)
plt.show()
```

```python
starting_node = 'O'

current_node = starting_node
remaining_nodes = list(G.nodes)
while remaining_nodes:
    remaining_nodes.remove(current_node)
    print(current_node)
```
